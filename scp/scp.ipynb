```python
import pandas as pd
import igraph as ig

from itertools import combinations
path = "../files/"
```

```python
regulome_df = pd.read_csv(path+"human_regulome_pd.gz").drop(columns='LLR_GRG')
regulome_df = regulome_df.astype({'ProteinAid': 'str', 'ProteinBid':'str'})
```

```python
# def scp(edge_df, k, weight):
k = 4

graph = ig.Graph.DataFrame(regulome_df, directed=True, use_vids=False) # maybe add difference directed and undirected - argument
regulome_df.sort_values(by = ['PPV'])
edges = graph.get_edgelist() # igraph adds and numebr links/edgs according to order in df. SO can assume that edges are already ordered

k_cliques = set()
graph.delete_edges(edges)
for edge in edges:
    # print(edge)
    vi, vj = edge[0], edge[1]
    graph.add_edge(vi, vj)
    # minimum requirement - both nodes vi and vj have degree of at least k-1
    if (graph.degree(vi) < (k-1)) or (graph.degree(vj) < (k-1)):
        continue
    neighbors = set(graph.neighbors(vi)).intersection(set(graph.neighbors(vj)))
    # print(vi, vj, neighbors)
    if len(neighbors) < k-2:
        continue
    
    new_cliques = set()
    for neighbor in neighbors:
        
        neighbors_n = set(graph.neighbors(neighbor)).intersection(neighbors)
        if len(neighbors_n) < (k-3):
            continue
        # print(neighbor, neighbors_n)
        for extras in combinations(neighbors_n, (k-3)):
            new_clique = set([vi, vj, neighbor, *[extra for extra in extras]])
            k_cliques.add(frozenset(new_clique))


 

```
